```python
import requests
import pandas as pd
import array as arr
from bs4 import BeautifulSoup
import numpy as np
```

```python
url = "https://www.loterias.com/baloto/resultados/2022"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
tabla = pd.DataFrame()
```

```python
encabezados = soup.find("thead").find_all("th")
fechas = soup.find("tbody").find_all("a")
results = soup.find("tbody").find_all("td", class_ = "centred baloto")

```

```python
encabezado = []
fecha = []
for e in encabezados:
    encabezado.append(e.text)

for f in fechas:
    fecha.append(f.text)

resulta = []
for n in results:
  itms = n.find_all("li", attrs={"ball"})
  for r in itms:
    resulta.append(r.text)

resultado1 = [resulta[i:i + 6] for i in range(0, len(resulta), 6)]
resultado2 = [resultado1[i:i + 2] for i in range(0, len(resultado1), 2)]

baloto = []
revancha = []
for l in resultado2:
  baloto.append(l[0])
  revancha.append(l[1])
```

```python
tabla[encabezado[0]] = fecha
tabla["Baloto"] = baloto
tabla["Revancha"] = revancha
tabla.head(100)
```
